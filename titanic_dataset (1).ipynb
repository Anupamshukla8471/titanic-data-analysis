```python
sns.heatmap(df_tips.corr(numeric_only="TRUE"))
# df_tips.corr(numeric_only="TRUE")

# Computes the correlation matrix only for numeric columns.
```

```python
sns.heatmap(df_tips.corr(numeric_only='TRUE'),cmap="coolwarm")
```

```python
df_tips["sex"]= df_tips["sex"].map({"Female":1,"Male":0})
df_tips
```

```python
df_tips["smoker"]=df_tips["smoker"].map({"No":0,"yes":1})
df_tips["day"]=df_tips["day"].map({"Thur":0,"Fri":1,"sat":3,"sun":4})
df_tips.time.map({"Lunch":0,"Dinner":1})
df_tips
```

```python
sns.heatmap(df_tips.corr(numeric_only="TRUE"),cmap="coolwarm",annot=True)
# linear regression y=mx+c
```
